```python
# Load the Dataset
import pandas as pd

df = pd.read_csv('movies (1).csv')  # Replace with the correct filename after upload
df.head()
```

```python
# Dataset Information
print("Shape of dataset:", df.shape)
print("Column names:", df.columns.tolist())
df.info()
df.describe(include='all')
```

```python
# Check for Missing Values and Duplicates
print("Missing Values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Extract Year from Title (if exists)
df['year'] = df['title'].str.extract(r'\((\d{4})\)').astype(float)

# Count how many genres per movie
df['genre_count'] = df['genres'].apply(lambda x: len(x.split('|')) if pd.notnull(x) else 0)
df.head()

```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

```python
# Visualize Most Common Genres
import matplotlib.pyplot as plt
import seaborn as sns

# Count genre appearances
from collections import Counter

all_genres = sum(df['genres'].dropna().str.split('|').tolist(), [])
genre_counts = Counter(all_genres)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=list(genre_counts.keys()), y=list(genre_counts.values()))
plt.xticks(rotation=45)
plt.title("Frequency of Movie Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Number of Movies Released Per Year
plt.figure(figsize=(12,6))
df['year'].value_counts().sort_index().plot(kind='bar')
plt.title("Number of Movies Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

```python
# Simple Search Function - Search by Genre
def search_by_genre(genre):
    return df[df['genres'].str.contains(genre, case=False, na=False)][['title', 'genres']]

# Example usage
search_by_genre('Comedy').head(10)
```
